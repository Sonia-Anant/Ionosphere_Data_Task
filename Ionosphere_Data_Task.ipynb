```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
```

```python
#loading the data set into dataframe
df = pd.read_csv("E:/Datasets/ionosphere.csv")
df.head(10)
```

```python
df.shape
```

```python
#getting detail of the dataframe
df.info()
```

```python
#checking missing values
print(df.isnull().sum())    
```

```python
# Rows containing duplicate data
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows:", duplicate_rows_df.shape)
```

```python
df.describe()
```

```python
#feature V2 seems to be redundant and we can drop it
df.drop('V2', axis=1, inplace=True)
```

```python
df.head()
```

```python
#defining X and y from our dataset
X = df.drop("Class",axis=1)
X
```

```python
y = df['Class']
```

```python
#generalize the dataset
from sklearn import preprocessing
X = preprocessing.MinMaxScaler().fit(X).transform(X)
X[0:5]
```

```python
#train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=10, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
predictions = LR.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
accuracy=confusion_matrix(y_test,predictions)
accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy_s =accuracy_score(y_test,predictions)
accuracy_s
```

```python
predictions
```

```python
from sklearn.tree import DecisionTreeClassifier 
DCT = DecisionTreeClassifier()
DCT.fit(X_train,y_train)
```

```python
predictions = DCT.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
accuracy=confusion_matrix(y_test,predictions)
accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy_s =accuracy_score(y_test,predictions)
accuracy_s
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

Ionopshere dataset contain "Class" as the feature label, containing binary values. Thus we can apply Classification Task under Supervised Learning. Applied Logistic Regression and Decision Tree Classifier, Logistic Regression is performing better.
